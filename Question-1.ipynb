```python
# Part 1 :

t=np.linspace(0,100,5000)
fig,axes=plt.subplots(2,1,figsize=(12,5)) 

def Input_current(t) :
    if (10<=t) & (t<=75) :
        return 2
    else :
        return 0 

I=list(map(Input_current,t))
axes[0].plot(t,I)
axes[0].set_title('Input Current v/s Time')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Input Current')

def Potential (t) :
    global t_fire

    I = Input_current(t) 
    R = 2 
    C = 5

    if (t>=t_fire -5) and (t<= t_fire) and (t_fire!=10) :
        return 0 

    u = R * I * (1 - np.exp( -(t - t_fire ) / (R*C) ) )
    
    if (u>=1) : 
        t_fire =t + 5

    return u 

t_fire=10 

U = list(map(Potential,t))
axes[1].plot(t,U)
axes[1].set_title('Membrane Potential v/s Time')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Membrane Potential')

plt.tight_layout()
```

```python
pi=3.14159265359

fig,axes=plt.subplots(2,1,figsize=(12,5)) 
t=np.linspace(0,100,50000)

def Input_current(t) :
    w=45*pi/180
    return 5*np.sin(w*t)


I=list(map(Input_current,t))
axes[0].plot(t,I)
axes[0].set_title('Input Current v/s Time')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Input Current')


def Potential (t) :
    global t_fire
    R = 2
    C = 5 
    y0=0

    y=np.zeros(len(t))
    y[0]=y0  

    for n in range(0,len(t)-1) :

        I=Input_current(t[n])
        f = lambda y,t :(-y + I * R) / (R * C) 
        y[n+1] = y[n] + f(y[n],t[n]) * (t[n+1]-t[n])

        if (y[n+1]>=1) : 
            t_fire =t[n] + 5

        if (t[n]>=t_fire -5) and (t[n]<= t_fire) and (t_fire!=0) :
            y[n+1]=0

    return y

t_fire=0

U=Potential(t)

axes[1].plot(t,U)
axes[1].set_title('Membrane Potential v/s Time')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Membrane Potential')

plt.tight_layout()
```
